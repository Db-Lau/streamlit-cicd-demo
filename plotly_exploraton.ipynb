```python
fig = px.line(nordic, 
              x = "year",
              y = "gdpPercap",
              color = "country",
              title = "Nordic countries GDP per capita 1952-2007")

fig.update_layout(hovermode = "x")

fig.update_xaxes(showspikes = True,
                 spikedash = "solid",
                 spikecolor = "green",
                 spikemode = "across",
                 spikesnap = "cursor")

fig.show()
```

```python
#bubble chart for showing more than 2 D

fig = px.scatter(gapminder, 
                 x = "gdpPercap",
                 y = "lifeExp",
                 size = "pop",
                 color = "country",
                 size_max = 70, 
                 animation_frame = "year",
                 #animation_group = "country",
                 title = "Gapminder animation",
                 range_x = [100, 10000])

fig.show()
```

```python
import pandas as pd
import numpy as np
```

```python
stocks = px.data.stocks()
stocks.head()



```

```python
fig = px.line(stocks,
              x = "date",
              y = "GOOG")

fig.show()
```

```python
stocks_long = stocks.melt(id_vars="date", var_name="Company", value_name="Price")
```

```python
stocks_long
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate uneven trading volume using a log-normal distribution
mean_log = 7  # Adjust for desired mean volume
sigma_log = 0.8  # Adjust for spread/skewness

stocks_long["Volume"] = np.random.lognormal(mean=mean_log, sigma=sigma_log, size=stocks_long.shape[0])

# Scale to realistic stock trading volumes
stocks_long["Volume"] *= 10  # Adjust scale factor as needed
```

```python
stocks_long
```

```python
google = stocks_long[stocks_long["Company"]=="GOOG"]

fig = px.line(google,
              x = "date",
              y = "Price")

fig.show()
```

```python
fig = px.line(stocks_long,
              x = "date",
              y = "Price",
              color = "Company")

fig.show()
```

```python
fig = px.scatter(stocks_long, 
                 x='Volume', 
                 y='Price', 
                 size='Volume',  # Bubble size represents trading volume
                 color='Company',  # Different colors for different stocks
                 title="Stock Prices and Trading Volume Over Time",
                 animation_frame="date")

fig.show()
```
